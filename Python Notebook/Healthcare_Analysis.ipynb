# Data Analysis 

## 1.Patient Demographics & Conditions

### *Q.1) Which medical conditions are most common among patients (e.g., Cancer, Obesity, Diabetes)?*

```python
data['Medical Condition'].value_counts().head(1)
```

### *Q.2) What is the average age of patients by medical condition? (e.g., Is cancer more common in older patients?)*

```python
data.groupby(['Medical Condition']) ['Age'].mean().reset_index().sort_values(by='Age',ascending= False)
```

### *Q.3) Is there any relationship between gender and medical condition occurrence?*

```python
data.groupby(['Medical Condition', 'Gender']).size().unstack(fill_value=0)
```

# 2. Hospital & Admission Insights

### *Q.4) Which admission type (Urgent, Emergency, Elective) is most frequent?*

```python
data['Admission Type'].value_counts()
```

### *   Q.5) Which hospitals handle the highest number of patients?*

```python
data['Hospital'].value_counts().head(10)
```

### *Q.6) Which doctors treat the most patients, and what conditions do they treat most often?*

```python
highest_patients = data[['Medical Condition','Doctor']].value_counts().reset_index(name= 'Total Patients')
highest_patients.head(10)
```

# 3. Financial & Insurance Insights

### *Q.7) Which medical conditions have the highest average billing amount?*

```python
data.groupby(['Medical Condition']) ['Billing Amount'].mean().reset_index(name= 'Avg Billing Amount').sort_values(by = 'Avg Billing Amount',ascending=False)
```

### *Q.8) Do patients with different insurance providers show differences in billing amounts? (e.g., Blue Cross vs Medicare vs Aetna)*

```python
data.groupby('Insurance Provider')['Billing Amount'].mean().sort_values(ascending=False)
```

# 4.Medication & Test Results

### *Q.9) Which medications are prescribed most frequently for each condition?*

```python
data['Medication'].value_counts().reset_index(name= 'Total Frequently')
```

### *Q.10) Is there a correlation between medical condition and test result type (Normal, Abnormal, Inconclusive)?*

```python
data.groupby(['Medical Condition','Medical Condition']).size().unstack(fill_value=0)
```
